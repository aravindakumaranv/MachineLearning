```python
import pandas as pd
```

```python
d = pd.read_csv('auto-mpg.csv')
d.origin.replace(
    to_replace={
        1: 'america',
        2: 'europe',
        3: 'japan'
    },
    inplace=True
)

d.origin = pd.Categorical(d.origin)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Visualizing one continuous variable

```python
fig, ax = plt.subplots(figsize=(8,8))
sns.histplot(data=d, x='mpg', ax=ax);
```

```python
bins = range(5,55,5)
fig, ax = plt.subplots(figsize=(8,8))
sns.histplot(data=d, x='mpg', ax=ax, bins=bins);
```

```python
bins = range(5,55,5)
fig, ax = plt.subplots(figsize=(8,8))
sns.histplot(data=d, x='mpg', ax=ax, bins=bins, kde=True);
ax.set_xticks(bins);
```

## Visualizing a categorical variable

```python
fig, ax = plt.subplots(figsize=(8,8))
sns.countplot(x=d['origin'], ax=ax);
```

## Visualize two continuous variables at the same time

```python
fig, ax = plt.subplots(figsize=(8,8))
sns.scatterplot(data=d, x='weight', y='mpg', ax=ax);
```

## Visualizing two continuous and one categorical variable at the same time

```python
fig, ax = plt.subplots(figsize=(8,8))
sns.scatterplot(data=d, x='weight', y='mpg', hue='origin', ax=ax);
```
